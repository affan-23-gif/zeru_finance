```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load scores + features
df_scores = pd.read_csv("../src/wallet_scores.csv")
df_features = pd.read_csv("../src/wallet_scores.csv")  # Can use same file or merged one

# Plot: Score distribution
plt.figure(figsize=(10, 5))
sns.histplot(df_scores['credit_score'], bins=10, kde=True)
plt.title("Wallet Credit Score Distribution")
plt.xlabel("Credit Score (0–1000)")
plt.ylabel("Wallet Count")
plt.grid(True)
plt.show()

# Bin into ranges
bins = list(range(0, 1100, 100))
labels = [f"{b}-{b+99}" for b in bins[:-1]]
df_scores['score_bin'] = pd.cut(df_scores['credit_score'], bins=bins, labels=labels, include_lowest=True)

# Count wallets in each bin
bin_counts = df_scores['score_bin'].value_counts().sort_index()
print("\nWallets per score bin:")
print(bin_counts)

```

Most wallets score between 0–200

There's a long tail of well-behaved wallets (score 600–1000)

This supports that risky/dead wallets dominate, and only a minority exhibit good DeFi behavior
